```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_excel('/content/adult update 2.xlsx')

# Filter for people with a Masters degree who are divorced
masters_divorced = df[(df['education'].str.strip() == 'Masters') & (df['marital_status'].str.strip() == 'Divorced')]

# Filter for people who did not finish high school and are divorced
no_hs_divorced = df[(df['education'].str.strip() == 'Did not Finish HS') & (df['marital_status'].str.strip() == 'Divorced')]

# Count the number of occurrences
counts = {
    'Masters & Divorced': len(masters_divorced),
    'Did not Finish HS & Divorced': len(no_hs_divorced)
}

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(counts.keys(), counts.values(), color=['blue', 'orange'])
plt.xlabel('Education Level & Marital Status')
plt.ylabel('Number of People')
plt.title('Comparison of Divorced People with Masters vs. Did not finish HS')
plt.show()
```
